# Sentiment Analysis of Social Media Data(Twitter as a case study)

```python
#import necessary libraries

import pandas as pd
from ntscraper import Nitter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
#create an object for the nitter

scraper = Nitter() #downloads some instances
```

```python
#import datetime to be able to work with dates.

from datetime import datetime, timedelta
```

```python
#scrape tweets from 10th August,2023 till 10th September,2023

tweets = scraper.get_tweets('jagunjagunmovie', mode='hashtag', since='2023-08-10', until='2023-09-10')
```

```python
tweets
```

```python
#break tweets by including '-----Next Tweet------- after each tweet'

for tweet in tweets['tweets']:
    print(tweet)
    print('------Next Tweet----')
```

```python
#create an empty list named final_tweet and create a list named data with link, user, username, text, date, is-retweet, likes, retweet.
#append data to final_tweets

final_tweets = []
for tweet in tweets["tweets"]:
    data = [
        tweet["link"],
        tweet["user"]["name"],
        tweet["user"]["username"],
        tweet["text"],
        tweet["date"],
        tweet["is-retweet"],
        tweet["stats"]["likes"],
        tweet["stats"]["retweets"],
    ]
    final_tweets.append(data)
```

```python
final_tweets
```

```python
#create a dataframe called data to include columns (link, name, username, text, date, retweeted, No_of_likes, No_of_retweets)

data = pd.DataFrame(
    final_tweets,
    columns=[
        "link",
        "name",
        "username",
        "text",
        "date",
        "retweeted",
        "No_of_likes",
        "No-of_retweets",
    ],
)
```

```python
#convert to csv

data.to_csv('twitter_data_final.csv')
```

```python
#import matplotlib for visualization

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
#Generate a wordcloud

import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Create an empty string to store the combined text from all rows
combined_text = ""

# Iterate through each row in the 'text' column and concatenate the text
for row in data['text']:
    combined_text += str(row) + " "

# Create and generate a word cloud image:
wordcloud = WordCloud(collocations=False, max_words=400, background_color='white').generate(combined_text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#convert wordcloud to image

wordcloud.to_file("wordcloud.png")
```
